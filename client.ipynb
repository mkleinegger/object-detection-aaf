### Post all

```python
import os
from yolo_model import load_image_as_base64
import requests


input_files_path = "input_data"
idx = []

for file_name in os.listdir(input_files_path):
    path_to_img = input_files_path + "/" + file_name
    img_data, uuid = load_image_as_base64(path_to_img)
    idx.append(uuid)

    post_data = {"id":str(uuid), "image_data":img_data}

    requests.post(f"http://127.0.0.1:5000/images/{uuid}",json=post_data)
    
```

### Get all

```python
for uuid in idx:
    response = requests.get(f"http://127.0.0.1:5000/images/{uuid}")
    if response.status_code == 200:
        print(response.json())
```
